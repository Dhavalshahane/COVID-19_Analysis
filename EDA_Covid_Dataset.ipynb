```python
import numpy as np
import pandas as pd
import matplotlib as plt
import seaborn as sns
```

```python
#UPLOAD DATASET
Covid_data = pd.read_csv(r'Covid_data.csv')
```

```python
# Perform NumPy operations on the dataset
import numpy as np

# Assuming 'Confirmed' is the column name for confirmed cases in your 'Covid_data'
confirmed_cases = Covid_data['Confirmed']  # Extract confirmed cases data
deaths = Covid_data['Deaths']  # Replace 'Deaths' with actual column name
recovered = Covid_data['Recovered']  # Replace 'Recovered' with actual column name

# Basic statistics using numpy
mean_confirmed = np.mean(confirmed_cases)
mean_deaths = np.mean(deaths)
mean_recovered = np.mean(recovered)

median_confirmed = np.median(confirmed_cases)
median_deaths = np.median(deaths)
median_recovered = np.median(recovered)

std_confirmed = np.std(confirmed_cases)
std_deaths = np.std(deaths)
std_recovered = np.std(recovered)

print("Mean Confirmed Cases:", mean_confirmed)
print("Mean Deaths:", mean_deaths)
print("Mean Recovered:", mean_recovered)

print("Median Confirmed Cases:", median_confirmed)
print("Median Deaths:", median_deaths)
print("Median Recovered:", median_recovered)

print("Standard Deviation of Confirmed Cases:", std_confirmed)
print("Standard Deviation of Deaths:", std_deaths)
print("Standard Deviation of Recovered:", std_recovered)
```

```python
#  PIECHART
import matplotlib.pyplot as plt # Import the pyplot submodule

# Calculate the proportion of confirmed cases for each WHO region
region_counts = Covid_data['WHO Region'].value_counts()

# Create the pie chart
plt.figure(figsize=(8, 8))
plt.pie(region_counts, labels=region_counts.index, autopct='%1.1f%%', startangle=90)
plt.title('Distribution of Confirmed Cases by WHO Region')
plt.axis('equal')  # Ensure the pie chart is a circle
plt.show()
```

```python
#BARGRAPH
import matplotlib.pyplot as plt # Import the pyplot module from matplotlib and alias it as 'plt'
import seaborn as sns # Import the seaborn module and alias it as 'sns'

# Distribution of countries by WHO Region
plt.figure(figsize=(10, 6))
sns.countplot(y='WHO Region', data=Covid_data, order=Covid_data['WHO Region'].value_counts().index)
plt.title('Distribution of Countries by WHO Region')
plt.show()
```

```python
import matplotlib.pyplot as plt

# Step 1: Identify the top 10 countries with the most deaths
top_deaths_countries = Covid_data[['Country/Region', 'Deaths']].sort_values(by='Deaths', ascending=False).head(10)

# Step 2: Plot a bar graph of the top 10 countries with the most deaths
plt.figure(figsize=(12, 6))
plt.bar(top_deaths_countries['Country/Region'], top_deaths_countries['Deaths'], color='red')
plt.xlabel('Country/Region')
plt.ylabel('Number of Deaths')
plt.title('Top 10 Countries with the Most Deaths')
plt.xticks(rotation=45)
plt.show()
```

```python
# Step 1: Identify the top 10 countries with the most recoveries
top_recovered_countries = Covid_data[['Country/Region', 'Recovered']].sort_values(by='Recovered', ascending=False).head(10)

# Step 2: Plot a bar graph of the top 10 countries with the most recoveries
plt.figure(figsize=(12, 6))
plt.bar(top_recovered_countries['Country/Region'], top_recovered_countries['Recovered'], color='green')
plt.xlabel('Country/Region')
plt.ylabel('Number of Recoveries')
plt.title('Top 10 Countries with the Most Recoveries')
plt.xticks(rotation=45)
plt.show()
```

```python
# BOXPLOT

plt.figure(figsize=(10, 6))
sns.boxplot(x='WHO Region', y='Confirmed', data=Covid_data)
plt.title('Boxplot of Confirmed Cases by WHO Region')
plt.xlabel('WHO Region')
plt.ylabel('Confirmed Cases')
plt.xticks(rotation=45)
plt.show()
```

```python
# Correlation matrix
# Select only numerical columns for correlation calculation
numerical_data = Covid_data.select_dtypes(include=['number'])
correlation_matrix = numerical_data.corr()

# Plotting the correlation matrix
plt.figure(figsize=(12, 8))
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm')
plt.title('Correlation Matrix')
plt.show()
```
